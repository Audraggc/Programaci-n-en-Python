##4.2 Llamando a una función

```
# Llamada a una función con un parámetro
def greet(name):
  print('Hello', name)

greet('Rosa')
```

```
# Definición de la función
def factorial(n):
    fact = 1
    for x in range(1, n + 1):
        fact = fact * x
    return fact

# La función combinatoria calcula el número de posibles
# combinaciones al seleccionar un número m de elementos 
# de un conjunto de n elementos
def combinatory(n, m):
    return factorial(n) / (factorial(m) * factorial(n-m))

# Llamada a la función
print(combinatory(6, 2))
```

```
# Definición de una función para calcular el área de un triángulo
def triangle_area(b, h):
    return b * h / 2

a = triangle_area(2.0, 5.0)
print('Área:', a)
```

```
# Contar el número de veces que escribió mi jefe en una semana
file_handler = open('/content/sample_data/email.txt', 'r')
count = 0
for line in file_handler:
    data = line.split(',')
    if data[0] == 'Pedro Gómez':
        count = count + 1
print('Recibí', count, 'correos de mi jefe durante esta semana.') 
file_handler.close()
```

```
# Contar el número de correos que recibí en diferentes
# momentos del día
file_handler = open('/content/sample_data/email.txt', 'r')
count = 0
dia = 0
tarde = 0
noche = 0
for line in file_handler:
    if not line.startswith('Remitente'):
        data = line.split(',')
        hora = int(data[2])
        if hora >= 8 and hora < 12: 
            dia = dia + 1 
        elif hora >= 12 and hora < 20: 
            tarde = tarde + 1 
        else:
            noche = noche + 1

print('Recibí', dia, 'correos durante el día.')
print('Recibí', tarde, 'correos durante la tarde.')
print('Recibí', noche, 'correos durante la noche.') 
file_handler.close()
```
