# **TEMA 3. Estatutos de repetición**

##3.1 While

```
# Estatuto while simple
numtxt = input('Dame un entero positivo')
num = int(numtxt)
i = 0
while i < num:
    print(i)
    i = i + 1

print('Primera línea fuera del while')
```

```
# Estatuto break
grade = 0
promedio = 0
i = 0
while i < 10:
    grade_txt = input('Dame una calificación o un número negativo para terminar')
    grade = float(grade_txt)
    print(grade)
    if grade < 0:
        break
    promedio = promedio + grade
    i = i + 1

print('Primera línea fuera del while')
promedio = promedio / i;
print('El promedio es:', promedio)
```

```
# Estatuto continue
promedio = 0
while i < 10:
    grade_txt = input('Dame una calificación o un número negativo para un NP')
    grade = float(grade_txt)
    print(grade)
    if grade < 0:
        continue
    promedio = promedio + grade
    i = i + 1

print('Primera línea fuera del while')
promedio = promedio / i;
print('El promedio es:', promedio)
```

##3.2 For

```
# Estatuto for iterando sobre una cadena de caracteres
txt = input('Dame un texto: ')
i = 0
for c in txt:
    i = i + 1
print('Número de caracteres:', i)
```

```
# Estatuto for iterando sobre una lista
lst = [10, 74, 23, 80, 9]
avg = 0
for n in lst:
    print(n)
    avg = avg + n
avg = avg / len(lst)
print('Número de elementos:', len(lst))
print('Promedio:', avg)
```

```
# Estatuto for iterando sobre rangos
# Iterar desde cero hasta un número antes del límite
x = range(4)
for n in x:
  print(n)
print()

# Iterar desde el número inicial hasta un número antes del límite
x = range(2, 10)
for n in x:
  print(n)
print()

# Iterar desde 2 hasta antes del 10 pero aumentando de dos en dos
# Imprime solo los números pares 
x = range(2, 10, 2)
for n in x:
  print(n)
```

# **TEMA 4. Definición de Funciones**

##4.1 Estructura de una función

```
# Definición de una función simple
def greet():
  print("Hello!")

greet()
```

```
# Definición de una función con un parámetro
def greet(name):
  print('Hello', name)

greet('Rosa')
```

```
# Definición de una función que tiene un parámetro
# y un valor por defecto para el mismo
def greet(name = 'World'):
  print('Hello', name)

greet('Rosa')
greet()
```

```
# Definición de una función que recibe dos parámetros
def add(a, b):
  return a + b

x = add(2, 5)
print("Addition: ", x)
```

##4.2 Llamando a una función

```
# Llamada a una función con un parámetro
def greet(name):
  print('Hello', name)

greet('Rosa')
```

```
# Definición de la función
def factorial(n):
    fact = 1
    for x in range(1, n + 1):
        fact = fact * x
    return fact

# La función combinatoria calcula el número de posibles
# combinaciones al seleccionar un número m de elementos 
# de un conjunto de n elementos
def combinatory(n, m):
    return factorial(n) / (factorial(m) * factorial(n-m))

# Llamada a la función
print(combinatory(6, 2))
```

```
# Definición de una función para calcular el área de un triángulo
def triangle_area(b, h):
    return b * h / 2

a = triangle_area(2.0, 5.0)
print('Área:', a)
```

***Archivos vacíos: os stat y st_size***

```
import os

# Revisar si un archivo está vacío
def file_is_not_empty(name):
  return os.stat(name).st_size > 0

file_name = '/content/sample_data/prueba.txt'

if file_is_not_empty(file_name):
  file_handler = open(file_name)
  for line in file_handler:
    print(line)
  file_handler.close()
else:
  print('The file is empty')
```

```
# Contar el número de renglones de un archivo
file_handler = open('/content/sample_data/prueba.txt', 'r')
count = 0
for line in file_handler:
    count = count + 1
print('Número total de líneas:', count)
file_handler.close()
```

## TEMA 5.2 Escribir en un archivo de texto

***Listado recursivo de archivos***

***Paquete glob***

```
import glob

# Obtener todos los archivos con una misma extensión
# dentro de un directorio
path = '/content/sample_data'
files = glob.glob(path + '/*.txt')
print(files)
```

***Recursivamente***

```
import glob

# Obtener todos los archivos con una misma extensión
# dentro de un directorio de forma recursiva
path = '/content/sample_data'
files = glob.glob(path + '/**/*.txt', recursive=True)
print(files)
```

***Iterador***

```
import glob

# Obtener todos los archivos con una misma extensión
# dentro de un directorio y guardarlos en un iterador
path = '/content/sample_data'
files = glob.iglob(path + '/*.txt')
for file in files:
  print(file)
```
