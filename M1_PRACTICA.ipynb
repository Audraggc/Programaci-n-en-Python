##Acción 3: Segmentar los datos

```
# Abrir un archivo de texto y recuperar línea por línea 
# Descartando el primer renglón usando un booleano
# Imprimir el nombre del planeta y su periodo rotacional
file_handler = open('/content/sample_data/planets.csv', 'r') 
first = True 
for line in file_handler: 
    if not first: 
        data = line.split(',')
        if data[0] != 'NA' and data[1] != 'NA':
            nombre = data[0]
            periodo_rotacional = int(data[1])
            print(nombre, ':', periodo_rotacional) 
    first = False
file_handler.close()
```
